# 03 — Model Training

Train 4 models (Logistic Regression, Random Forest, XGBoost, LightGBM)
with RandomizedSearchCV hyperparameter tuning.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
from pathlib import Path
```

## 1. Load Processed Data

```python
processed = Path('../data/processed')

# Diabetes
X_train_d = pd.read_csv(processed / 'diabetes_X_train.csv')
X_test_d = pd.read_csv(processed / 'diabetes_X_test.csv')
y_train_d = pd.read_csv(processed / 'diabetes_y_train.csv').squeeze()
y_test_d = pd.read_csv(processed / 'diabetes_y_test.csv').squeeze()

# Heart
X_train_h = pd.read_csv(processed / 'heart_X_train.csv')
X_test_h = pd.read_csv(processed / 'heart_X_test.csv')
y_train_h = pd.read_csv(processed / 'heart_y_train.csv').squeeze()
y_test_h = pd.read_csv(processed / 'heart_y_test.csv').squeeze()

print(f'Diabetes — Train: {X_train_d.shape}, Test: {X_test_d.shape}')
print(f'Heart — Train: {X_train_h.shape}, Test: {X_test_h.shape}')
```

## 2. Train All Models — Diabetes

```python
from src.model import train_all_models, save_model

print('Training all models on Diabetes dataset...')
print('This uses RandomizedSearchCV with 5-fold StratifiedKFold.\n')

diabetes_results = train_all_models(X_train_d, y_train_d, n_iter=20)

for name, res in diabetes_results.items():
    print(f'{name}: best params = {res["best_params"]}')
    save_model(res['model'], name, 'diabetes')

print('\nAll diabetes models saved to models/')
```

## 3. Train All Models — Heart Disease

```python
print('Training all models on Heart Disease dataset...\n')

heart_results = train_all_models(X_train_h, y_train_h, n_iter=20)

for name, res in heart_results.items():
    print(f'{name}: best params = {res["best_params"]}')
    save_model(res['model'], name, 'heart')

print('\nAll heart models saved to models/')
```

## 4. Quick Validation

Verify models can predict on test data.

```python
from src.model import load_model

for name in diabetes_results:
    model = load_model(name, 'diabetes')
    pred = model.predict(X_test_d[:5])
    print(f'{name}: sample predictions = {pred}')
```

### 2.1 Identify Impossible Zero Values

In this dataset, 0 in Glucose, BloodPressure, SkinThickness, Insulin, and BMI
represents **missing data**, not actual measurements. A patient cannot have 0 blood pressure.

```python
zero_cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
zero_counts = (df_diabetes[zero_cols] == 0).sum()
zero_pct = (zero_counts / len(df_diabetes) * 100).round(1)

pd.DataFrame({'Zero Count': zero_counts, 'Percentage': zero_pct})
```

### 2.2 Target Distribution (Class Imbalance)

```python
fig, ax = plt.subplots(figsize=(6, 4))
df_diabetes['Outcome'].value_counts().plot(kind='bar', color=['#4e79a7', '#e15759'], ax=ax)
ax.set_title('Diabetes Outcome Distribution')
ax.set_xlabel('Outcome (0=No, 1=Yes)')
ax.set_ylabel('Count')
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
plt.tight_layout()
plt.show()

print(df_diabetes['Outcome'].value_counts())
print(f'\nImbalance ratio: {df_diabetes["Outcome"].value_counts()[0] / df_diabetes["Outcome"].value_counts()[1]:.2f}:1')
```

### 2.3 Feature Distributions

```python
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
for i, col in enumerate(df_diabetes.columns[:-1]):
    ax = axes[i // 4, i % 4]
    df_diabetes.boxplot(column=col, by='Outcome', ax=ax)
    ax.set_title(col)
    ax.set_xlabel('')
fig.suptitle('Feature Distributions by Diabetes Outcome', fontsize=14)
plt.tight_layout()
plt.show()
```

### 2.4 Correlation Heatmap

```python
fig, ax = plt.subplots(figsize=(10, 8))
corr = df_diabetes.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='RdBu_r',
            center=0, square=True, ax=ax)
ax.set_title('Diabetes Dataset — Correlation Matrix')
plt.tight_layout()
plt.show()
```

## 3. Heart Disease Dataset — Overview

```python
df_heart.head(10)
```

```python
df_heart.describe()
```

### 3.1 Target Distribution

```python
fig, ax = plt.subplots(figsize=(6, 4))
df_heart['target'].value_counts().plot(kind='bar', color=['#4e79a7', '#e15759'], ax=ax)
ax.set_title('Heart Disease Target Distribution')
ax.set_xlabel('Target (0=No Disease, 1=Disease)')
ax.set_ylabel('Count')
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
plt.tight_layout()
plt.show()

print(df_heart['target'].value_counts())
```

### 3.2 Feature Distributions

```python
numeric_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

fig, axes = plt.subplots(1, 5, figsize=(20, 4))
for i, col in enumerate(numeric_cols):
    df_heart.boxplot(column=col, by='target', ax=axes[i])
    axes[i].set_title(col)
    axes[i].set_xlabel('')
fig.suptitle('Feature Distributions by Heart Disease Target', fontsize=14)
plt.tight_layout()
plt.show()
```

### 3.3 Correlation Heatmap

```python
fig, ax = plt.subplots(figsize=(12, 10))
corr = df_heart.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='RdBu_r',
            center=0, square=True, ax=ax)
ax.set_title('Heart Disease Dataset — Correlation Matrix')
plt.tight_layout()
plt.show()
```
